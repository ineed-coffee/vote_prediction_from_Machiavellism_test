```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
filepath='D:/Dacon_Psychological_disposition/'
train = pd.read_csv(filepath+'train.csv')
test = pd.read_csv(filepath+'test_x.csv')
```

```python
train.shape
```

```python
test.shape
```

```python
train.columns
```

```python
# 출력 데이터를 제외한 나머지 컬럼이 모두 동일한가 = okay
(train.columns.drop('voted') != test.columns).sum()
```

```python
# 분할 지점 저장해두고 데이터 통합 후 전처리
split_point = train.shape[0]
data = pd.concat([train.drop('voted',axis=1),test],ignore_index=True)
```

```python
data.shape
```

```python
data.dtypes.value_counts() # 대부분 양적 데이터
```
